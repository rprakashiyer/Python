### Stacks muktiple dfs along 0/1 axis
Merge combines df in a smarter way based on values of columns

```python
pd.merge?
```

```python
users = pd.DataFrame({"userid":[1,2,3],"name":["sarath","Shruti","Khushal"]})
```

```python
users
```

```python
msgs = pd.DataFrame({"userid":[1,1,2,4],"msg":["hmmm","acchaa","theek hei","nice"]})
msgs
```

```python
users.merge(msgs,on="userid")
```

```python
users.merge(msgs,on="userid",how="left")
```

```python
users.merge(msgs,on="userid",how="outer")
```

```python
users.merge(msgs,on="userid",how="right")
```

```python
users.merge(msgs,on="userid",how="cross")
```

```python
users.rename([columns={"userid":"id"},inplace=True])
```
