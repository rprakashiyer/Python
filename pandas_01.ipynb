### If we have to drop some column, we use Drop function, with axis parameter
axis = 0 (Vertical Rearrangement Operation)
axis = 1 (Horizontal Rearrangement Operation)

```python
df.drop("Continent")
```

### So the default axis is 0 i.e Vertical rearrangement operation. If we have to search for a column and delete it, try to understand in this way. After deleting of the column, the rearrangment would happen horizontally or vertically. And basis that, then decide on the axis parameter. Because, we will be dropping a column, after dropping the rearrangement would happen horizontally hence axis = 1 

```python
df.drop("Continent",axis=1)
```

### But drop has not permanently affected the change

```python
df
```

```python
df.drop("Continent",axis=True,inplace=True)
```

```python
df
```
