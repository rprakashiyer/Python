### Vectorize

```python
np.vectorize
```

### Vectorize is a function which acts similar to that of a map function 

```python
a = np.arange(1,13).reshape(4,3)
a
```

### Here a function calculated is created, and this function could be as complicated as possible. It is then mapped through vectorize function of numpy to a variable. And here we are, we can use this variable now on any numpy array

```python
def calculated(var):
  return var**3

func1 = np.vectorize(calculated)

b = func1(a)
b
```

### Here the memory location of both a & b are same, But if we try to do any operation on the array, then the memory location would change. The numpy header will also be updated with new location of the variable

```python
a = np.arange(1,13)
a
```

```python
b = a.reshape(4,3)
```

### np.shares_memory() checks the header of the numpy array. And if both the header contents are matched then it returns True, else False. Here, the location of header is same, and the location pointed by headers in a & b are also same.

```python
np.shares_memory(a,b)
```

# 3D Matrices

Screenshot 2022-08-29 075225.png

```python
a=np.arange(1,25).reshape(2,3,4)
a
```

### In 3D numpy array When axis=0, its the depth on which the operation will be performed. Hence, addition is performed on 1st and 2nd layer

```python
np.sum(a,axis=0)
```

### when axis = 1, its the vertical operation in every layer

```python
np.sum(a,axis=1)
```

### when axis = 2, its the Horizontal operation in every layer

```python
np.sum(a,axis=2)
```

### Now we will see how splitting works
