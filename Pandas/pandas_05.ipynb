```python
movies = pd.read_csv('/content/movies.csv', index_col=0)
directors = pd.read_csv('/content/directors.csv',index_col=0)
data = movies.merge(directors, how='left', left_on='director_id',right_on='id')  
data.drop(['director_id','id_y'],axis=1,inplace=True)
```

```python
data
```

```python
data['budget'] = (data['budget']/1000000).round(2)
data['revenue'] = (data['revenue']/1000000).round(2)
data
```

### Transform works on one particular column

```python
def calc_exp(x):
  x = x - x.mean()
  x['budget']

data.groupby("director_name").transform(calc_exp)
data
```

### hence we need to select the column we need to work and then perform the function on that column

```python
def calc_exp(x):
  x = x - x.mean()
  return x

data["exp_or_inexp"] =data.groupby("director_name")['budget'].transform(calc_exp)
data
```

```python
data.shape
```

### Using apply() we can pass the dataframe, and work with multiple columns and perform the operation inside the function

```python
def calc_risky(x):
  x["risky_status"] = (x["budget"] - x["revenue"].mean()) >= 0
  return x

data_risky=data.groupby("director_name").apply(calc_risky)
data_risky
```

```python
data_risky[data_risky["risky_status"]==True][["title","director_name"]]
```

```python
data_risky.loc[data_risky["risky_status"]==True].head(10)
```
