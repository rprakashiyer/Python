### Using aggregate function we can apply multiple pre-defined functions to multiple columns at the same time

```python
data.groupby("director_name")[["title","year"]].aggregate({"title":"count","year":"min"})
```

```python
data.groupby("director_name")[["title","year"]].aggregate({"title":"count","year":"max"})
```

```python
df_agg = data.groupby("director_name")[["title","year"]].aggregate({"title":"count","year":["min","max"]})
```

### To add a new col call active years , which is max year minus min year

```python
df_agg["years_active"] = df_agg["year_max"] - df_agg["year_min"]
```

```python
df_agg
```

```python
df_agg["movies_per_year"] = df_agg["title_count"] / df_agg["years_active"]
```
